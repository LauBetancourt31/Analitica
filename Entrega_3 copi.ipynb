## <font color='00913f'> Categorización de variables
Justificación de la necesidad de categorizar variables y categorización si es necesaria

### Categorización de `GPSSpeed`

```python
df['GPSSpeed'].describe()
```

```python
# Calcula la media de GPSSpeed
media_vel = df['GPSSpeed'].mean()

# Calcula la diferencia entre la velocidad y la media
diferencia = df['GPSSpeed'] - media_vel

# Define los umbrales para "alto" y "bajo" en función de 0.06 de la media
umbral_alto = 0.06
umbral_bajo = - 0.06

# Crea la variable categórica según las condiciones dadas
df['Velocidad'] = pd.cut(diferencia, bins=[-float("inf"), umbral_bajo, umbral_alto, float("inf")], labels=['Rápida', 'Intermedia', 'Lenta'])

# Muestra el DataFrame resultante
df.head()
```

### Categorización de `ExhaustTemp_1`

```python
df['ExhaustTemp_1'].describe()
```

```python
# Calcula la media de ExhaustTemp_1
media_temp = df['ExhaustTemp_1'].mean()

# Calcula la diferencia entre la velocidad y la media
diferencia_temp = df['ExhaustTemp_1'] - media_temp

# Define los umbrales para "alto" y "bajo"
umbral_alto_temp = 45
umbral_bajo_temp = - 45

# Crea la variable categórica según las condiciones dadas
df['T_Escape'] = pd.cut(diferencia, bins=[-float("inf"), umbral_bajo_temp, umbral_alto_temp, float("inf")], labels=['Baja', 'Moderada', 'Alta'])

# Muestra el DataFrame resultante
df.head(20)
```

#### Prueba estadística

##### Rango intercuartílico

```python
import pandas as pd

umbral = 1.5
info = []

for variable in df_n:
    Q1 = df.describe().at['25%', variable]
    Q3 = df.describe().at['75%', variable]
    RIC = Q3 - Q1

    Max = Q3 + (umbral * RIC)
    Min = Q1 - (umbral * RIC)

    outliers = df[(df[variable] < Min) | (df[variable] > Max)]

    info.append([variable, Min, Q1, Q3, Max, RIC, len(outliers)])

ric = pd.DataFrame(info, columns=['Variable', 'Min', 'Q1', 'Q3', 'Max', 'RIC', 'Cantidad de Outliers'])
ric['Porcentaje de Outliers'] = (ric['Cantidad de Outliers'] / len(df)) * 100
ric
```
